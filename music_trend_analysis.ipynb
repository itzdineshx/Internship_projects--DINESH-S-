# **Music_trend_analysis**

**Description**:
genres are most popular but also what technologies are being used, what formats are selling the most, and which apps are creating viral moments for music.

```python
#importing libraries
import pandas as pd #for working with df
from sklearn.tree import DecisionTreeClassifier#for making Predictions
from sklearn.model_selection import train_test_split#for train the data to model
from sklearn.metrics import accuracy_score#check the accuracy score of model
import joblib#for saving the model
from sklearn import tree #for visualizing the decision tree
```

```python
#opening the music file
music_data = pd.read_csv('music.csv')
music_data
```

splitting the data

```python
x=music_data.drop(columns=['genre'])
x
```

```python
y=music_data['genre']
y
```

# Train the model

splitting the model

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

make the model to predict by Decision Tree

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
```

checking the accuracy score

```python
acc_score=accuracy_score(y_test,predictions)
acc_score
```

saving and loading the trained model

```python
joblib.dump(model,'music-recommender.joblib')
joblib.load('music-recommender.joblib')
predictions=model.predict([[21,1]])
predictions
```

visualizing the decision tree

```python
tree.export_graphviz(model,out_file='music-recommender.dot',
                    feature_names=['age','gender'],
                    class_names=sorted(y.unique()),
                    label='all',
                    rounded=True,
                    filled=True)
```
